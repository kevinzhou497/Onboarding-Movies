```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 


df = pd.read_csv('features_norm.csv')
df_features = df[['imdb_score_scaled', 'gross_scaled']]
df_features = df_features.sort_values(by=['gross_scaled'])



df_features.dropna(subset=['gross_scaled'], inplace=True)
df_features.dropna(subset=['imdb_score_scaled'], inplace=True)


X = np.array(df_features['imdb_score_scaled']).reshape(-1, 1) 
y = np.array(df_features['gross_scaled']).reshape(-1, 1) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0) 

 
lin_reg = LinearRegression()
lin_reg.fit(X, y)



#y_pred = regr.predict(X_test) 
#plt.scatter(X_test, y_test, color ='b') 
#plt.plot(X_test, y_pred, color ='k') 
  
#plt.show() 

plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('imdb_score_scaled')
plt.ylabel('gross_scaled')
plt.show()

print("Slope:" + str(lin_reg.coef_))
print("Intercept: " + str(lin_reg.intercept_))
print("Test R^2: " + str(lin_reg.score(X_test, y_test)))
print("Train R^2: " + str(lin_reg.score(X_train, y_train)))

Z, hmx, hmy = np.histogram2d(df_features['imdb_score_scaled'],df_features['gross_scaled'])

plt.title('Heatmap')
plt.xlabel('imdb_score_scaled')
plt.ylabel('gross_scaled')
plt.pcolormesh(hmx,hmy,Z.T)

plt.show()
```

```python
df = pd.read_csv('features_norm.csv')
df = df.sort_values('imdb_score_scaled')
df_features = df[['imdb_score_scaled', 'gross_scaled']]


df_features.dropna(subset=['gross_scaled'], inplace=True)
df_features.dropna(subset=['imdb_score_scaled'], inplace=True)


X = np.array(df_features['imdb_score_scaled']).reshape(-1, 1) 
y = np.array(df_features['gross_scaled']).reshape(-1, 1) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0) 
poly_reg = PolynomialFeatures(degree=5)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)

print(lin_reg.score(X_test, y_test)) 
# Visualizing the Polymonial Regression results
def viz_polymonial():
    plt.scatter(X, y, color='blue')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='red')
    plt.title('Polynomial Regression')
    plt.xlabel('imdb_score_scaled')
    plt.ylabel('gross_scaled')
    plt.show()
    return
viz_polymonial()

print("Slope:" + str(lin_reg.coef_))
print("Intercept: " + str(lin_reg.intercept_))
print("Score: " + str(lin_reg.score(X_test, y_test)))

```
